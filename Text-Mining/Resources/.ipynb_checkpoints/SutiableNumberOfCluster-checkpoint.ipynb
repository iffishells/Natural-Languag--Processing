```python
import  pandas as pd

corpus = pd.read_csv('dataset-CalheirosMoroRita-2017.csv', encoding = 'Latin-1', delimiter = '\n')
corpus.head(10)

docs = [ docs  for docs in corpus["Review"]]  ## getting single doc from the columns of Review 


from sklearn.feature_extraction.text import TfidfVectorizer
vec = TfidfVectorizer()
matrix_X = vec.fit_transform(docs)
matrix_X

matrix_X.toarray()  ## vector representation

from sklearn.cluster import KMeans
kn1 = KMeans(n_clusters=1)
kn1.fit(matrix_X)
print("kn1 : ",kn1.inertia_)

kn2 = KMeans( n_clusters=2)
kn2.fit(matrix_X)
print("kn2 : ",kn2.inertia_)

kn3 = KMeans( n_clusters=3)
kn3.fit(matrix_X)
print("kn3 : ",kn3.inertia_)

kn4 = KMeans( n_clusters=4)
kn4.fit(matrix_X)
print("kn4 : ",kn4.inertia_)

kn5 = KMeans( n_clusters=5)
kn5.fit(matrix_X)
print("kn5 : ",int(kn5.inertia_))



```

```python
import  matplotlib.pyplot as plt
list_inertia = [kn1.inertia_ , kn2.inertia_ ,kn3.inertia_ ,kn4.inertia_ ,kn5.inertia_]  ## list of inertia
no = [1,2,3,4,5]


## going to automation
list_inertia = []
no = []
for i in range(1,10):
    knn1 = KMeans(n_clusters=i)
    knn1.fit(matrix_X)
    list_inertia.append(knn1.inertia_)
    no.append(i)

plt.figure(figsize=(5,5))
plt.plot(no,list_inertia,marker = "o",markerfacecolor="Red")
plt.title("Finding the value of k ")
plt.xlabel("Number of cluster")
plt.ylabel("Inertia of cluster")
plt.show()
```
